```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg")

# Converting color image to grayscale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(gray,cmap='gray')

#resize
print("Original Dimension", gray.shape)
scale = 0.7
width = int((gray.shape[1])* scale)
height = int((gray.shape[0])* scale)
re_img = cv2.resize(gray,(width,height))
plt.subplot(1,2,2)
plt.title('Resized')
plt.imshow(re_img,cmap='gray')
print("Resized Dimension", re_img.shape)
plt.show()

#3D
X, Y = np.mgrid[0:re_img.shape[0], 0:re_img.shape[1]]
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, re_img , cmap = 'gray')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
