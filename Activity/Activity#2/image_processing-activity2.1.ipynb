```python
import cv2
import matplotlib.pyplot as plt
```

#2.1

```python
image = cv2.imread("harry.jpg")

scale = 0.5
width = int((image.shape[1])* scale)
height = int((image.shape[0])* scale)
re_img = cv2.resize(image,(width,height))

img_rgb = cv2.cvtColor(re_img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(re_img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(re_img, cv2.COLOR_BGR2HLS)
img_YCrCb = cv2.cvtColor(re_img, cv2.COLOR_BGR2YCrCb)

r,g,b = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]
h,s,v = img_hsv[:,:,0], img_hsv[:,:,1], img_hsv[:,:,2]
h1,l,s1 = img_hls[:,:,0], img_hls[:,:,1], img_hls[:,:,2]
y,cb,cr = img_YCrCb[:,:,0], img_YCrCb[:,:,1], img_YCrCb[:,:,2]

fig, ax = plt.subplots(4, 4, figsize=(12,12))
fig.patch.set_facecolor('white')

#RGB
ax[0, 0].set_title('RGB')
ax[0, 0].imshow(img_rgb)

ax[0, 1].set_title('R')
ax[0, 1].imshow(r,cmap ='gray')

ax[0, 2].set_title('G')
ax[0, 2].imshow(g,cmap ='gray')

ax[0, 3].set_title('B')
ax[0, 3].imshow(b,cmap ='gray')

#HSV
ax[1, 0].set_title('HSV')
ax[1, 0].imshow(img_hsv)

ax[1, 1].set_title('H')
ax[1, 1].imshow(h,cmap ='gray')

ax[1, 2].set_title('S')
ax[1, 2].imshow(s,cmap ='gray')

ax[1, 3].set_title('V')
ax[1, 3].imshow(v,cmap ='gray')

#HLS
ax[2, 0].set_title('HLS')
ax[2, 0].imshow(img_hls)

ax[2, 1].set_title('H')
ax[2, 1].imshow(h1,cmap ='gray')

ax[2, 2].set_title('L')
ax[2, 2].imshow(l,cmap ='gray')

ax[2, 3].set_title('S')
ax[2, 3].imshow(s1,cmap ='gray')

#YCrCb
ax[3, 0].set_title('YCrCb')
ax[3, 0].imshow(img_YCrCb)

ax[3, 1].set_title('Y')
ax[3, 1].imshow(y,cmap ='gray')

ax[3, 2].set_title('Cr')
ax[3, 2].imshow(cr,cmap ='gray')

ax[3, 3].set_title('Cb')
ax[3, 3].imshow(cb,cmap ='gray')

plt.show()
```
