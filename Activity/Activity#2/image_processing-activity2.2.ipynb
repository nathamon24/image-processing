```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("threebear.jpg")
img2 = cv2.imread("harry.jpg")

center = (img1.shape[1]/2,img1.shape[0]/2)
img1_crop = cv2.getRectSubPix(img1, img2.shape[:2][::-1],center)
```

```python
size = img2.shape
def frame(img1,img2):
  img_array = []
  for i in range(201):
    w1 = np.full(size, i*0.005)
    w2 = np.ones(size)
    w2 -= w1
    img = (w1*img1) + (w2*img2)
    img = img.astype(np.uint8)
    img_array.append(img)
  return img_array
```

```python
a = frame(img1_crop,img2)
b = frame(img2,img1_crop)
result = a + b
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
writer = cv2.VideoWriter("result-2.2.mp4", fourcc, 30, img2.shape[:2][::-1])
for frame in range(len(result)):
  writer.write(result[frame])
writer.release()
```
