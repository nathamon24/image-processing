```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#2.3

```python
image = cv2.imread("harry.jpg")

#Reduced size
scale = 0.5
width = int((image.shape[1])* scale)
height = int((image.shape[0])* scale)
re_img = cv2.resize(image,(width,height))

#BGR to RGB
img_rgb = cv2.cvtColor(re_img, cv2.COLOR_BGR2RGB)

#create mask
mask = np.zeros(img_rgb.shape, dtype=np.uint8)

start_point = (125, 240)
end_point = (375, 50)
color = (255,255,255)
thickness = -1
mask = cv2.rectangle(mask, start_point, end_point, color, thickness)

#Bitwise AND
result = cv2.bitwise_and(img_rgb,mask)

fig, ax = plt.subplots(1, 3, figsize=(10,3))

ax[0].set_title("Original")
ax[0].imshow(img_rgb)

ax[1].set_title("Image Mask")
ax[1].imshow(mask,cmap="gray")

ax[2].set_title("Bitwise_AND() result")
ax[2].imshow(result)

plt.tight_layout()
```
